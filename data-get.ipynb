```python
import yfinance as yf
import pandas as pd
```

```python
tickers = [
    'ICICIBANK.NS',   
    'RELIANCE.NS',    
    'TCS.NS',         
    'GOLDBEES.NS',    
    'MUTHOOTFIN.NS',
    'BTC-USD',
    'USDINR=X',
    'CL=F'
]
```

```python
start_date = '2015-01-01'
end_date = '2025-03-01'
```

```python
data = yf.download(tickers, start=start_date, end=end_date)
```
